```python
# TA-lib used to recognise candlestick patterns
candle_names = talib.get_function_groups()['Pattern Recognition']
candle_names
```

```python
# favorite candles
candle_names = [
    'CDLSHOOTINGSTAR',
    'CDLPIERCING',
    'CDLDARKCLOUDCOVER',
    'CDLHARAMI',
    'CDLHAMMER',
    'CDLDOJI',
    'CDLENGULFING',
    'CDLMORNINGSTAR',
    'CDLEVENINGSTAR'
]

# candle_ranks_top2 = {
#     'CDLEVENINGSTAR': 1,
#     'CDLMORNINGSTAR': 2,
# }
```

```python
# read data
df = pd.read_csv('./resources/google_10yrs_ohlc_labelled.csv', index_col=0,parse_dates=True)
df
```

```python
# create columns for each pattern
for candle in candle_names:
    df[candle] = getattr(talib, candle)(df['Open'], df['High'], df['Low'], df['Close'])
```

```python
df
```

```python
df['candlestick_pattern'] = 'NO_PATTERN'
#shapes = []
annotations = []

for index, row in df.iterrows():
    for candle in candle_names:
        try:
            if row[candle] != 0:
                df.loc[index,'candlestick_pattern'] = candle
                #shapes.append(dict(x0=index, x1=index, y0=df.loc[index,'High'], y1=df.loc[index,'High'], line_width=2))
                annotations.append(dict(x=index, y=df.loc[index,'High'], xanchor='left', text=candle, textangle=-90, height=8))
        except:
            print(f'an exception occurred')

# clean up columns
df.drop(['Volume', 'Adj Close'], axis = 1, inplace = True)
df.drop(candle_names, axis = 1, inplace = True)
df
```

```python
df['candlestick_pattern'].value_counts()
```

```python
# candlestick plot
fig = go.Figure(data=[go.Candlestick(
    x=df.index, 
    open=df['Open'], 
    high=df['High'], 
    low=df['Low'], 
    close=df['Close']
)])

fig.update_layout(
    title='Google',
    yaxis_title='Price [USD]',
    #shapes = shapes,
    annotations=annotations
)

fig.show()
```

```python
# transform data from ohlc to culr (close, upper-shadow, lower-shadow, real-body)
# for falling real-body the 3 features are -ve

for index, row in df.iterrows():
    
    df.loc[index,'real_body'] = df.loc[index,'Close'] - df.loc[index,'Open']
    
    if df.loc[index,'Close'] >= df.loc[index,'Open']:   
        df.loc[index,'upper_shadow'] = df.loc[index,'High'] - df.loc[index,'Close']
        df.loc[index,'lower_shadow'] = df.loc[index,'Open'] - df.loc[index,'Low']    
    else:
        df.loc[index,'upper_shadow'] = df.loc[index,'Open'] - df.loc[index,'High']
        df.loc[index,'lower_shadow'] = df.loc[index,'Low'] - df.loc[index,'Close']  
        
# tidy up
#df.drop(['Open', 'High', 'Low'], axis = 1, inplace = True)
df = df.rename(columns={'Open': 'open', 'High': 'high', 'Low': 'low', 'Close': 'close'})
df['date'] = df.index
df = df[['date', 'open', 'high', 'low', 'close', 'upper_shadow', 'lower_shadow', 'real_body', 'candlestick_pattern']]

df
```

```python
# data to csv
df.to_csv('./resources/google_10yrs_culr_labelled.csv', index=False,)
```
